```python
import pandas as pd
```

```python
df = pd.read_parquet(r'D:\programming\python\Math_prof\Data\umath-test-00000-of-00001 (1).parquet')
df.head()
```

```python
print(df['subject'].value_counts())
```

```python
df_text = df[df["has_image"] == False]
```

```python
subjects = [
    "differential_calc",
    "integral_calc",
    "algebra",
    "multivariable_calculus",
    "precalculus_review",
    "sequences_series"
]

# Sample 4 questions from each subject (without replacement)
sampled_df = (
    df_text[df_text["subject"].isin(subjects)]
    .groupby("subject", group_keys=False)
    .apply(lambda x: x.sample(n=4, random_state=42))
    .reset_index(drop=True)
)

```

```python
sampled_df.to_csv("u_math_eval_subset.csv", index=False)

```
